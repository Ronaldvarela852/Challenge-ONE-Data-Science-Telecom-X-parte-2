##Analisis grafico y de correlación

```python
import plotly.express as px
px.histogram(datos_codificados, x='Churn', text_auto=True)
```

```python
datos_codificados['OnlineSecurity'].value_counts()
```

```python
matriz_correlacion = datos_codificados.corr()
```

```python
matriz_correlacion = matriz_correlacion.drop('Churn', axis=0)
```

```python
plt.figure(figsize=(20, 16))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

```python
# umbral de correlación
umbral = 0.75

# pares de variables altamente correlacionadas
correlacion_alta = set()
for i in range(len(matriz_correlacion.columns)-2):
    for j in range(i):
        if abs(matriz_correlacion.iloc[i, j]) > umbral:
            colname_i = matriz_correlacion.columns[i]
            colname_j = matriz_correlacion.columns[j]
            correlacion_alta.add((colname_i, colname_j))

print("Pares de variables altamente correlacionadas:")
for pares in correlacion_alta:
    print(pares)
```

```python
plt.figure(figsize=(20, 16))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```
